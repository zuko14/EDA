```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
## plotting using matplotlib
# plt scatter
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('graph in 2D')
plt.savefig('exMl.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'y*',linestyle='dashed',linewidth=2,markersize=20)
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('2D')
```

```python
## creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
```

```python
# sine 
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title('sine wave')

plt.plot(x,y,'b')
plt.show()
```
