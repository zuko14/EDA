```python
df=sns.load_dataset('tips')
```

 ## corelation with heatmap

```python
sns.heatmap(df.corr())
```

```python
            # jointplot for bivarianat analysis (more than 2 features)

sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
 # pair plot note:- to use it the feature need to be int or float value

sns.pairplot(df)

```

```python
sns.pairplot(df,hue='smoker')
```

```python
sns.pairplot(df,hue='sex')
```

               DIST PLOT

```python
sns.distplot(df['tip'])
```

 CATEGORICAL PLOTS
 - box plot
 - violinplot
 - countplot
 - bar plot

```python
# count plot

sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y='smoker',data=df)
```

```python
 # bar plot

sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(y='tip',x='sex',data=df)
```

```python
sns.boxplot(x='total_bill',y='sex',data=df)
```

```python
sns.boxplot(x='day',y='tip',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```
